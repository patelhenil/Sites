```python
import pandas as pd
import geopandas as gp
import fiona
from shapely.geometry import shape,mapping, Point, Polygon, MultiPolygon
import csv
```

```python
df = gp.read_file('sites.geojson')
df
```

```python
with open('Fuel_Locations_Heartland.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    next(readCSV)
    for row in readCSV:
        lat = float(row[29])
        lon = 0 - float(row[30]) 
        point = Point(lon, lat)
        #print(point)
        ans = df['geometry'].contains(point)
        name = df['disp_name']
        if(sum(ans) != 0):
            print(f'{row[0]},{df['disp_name']},True')
        else:
            print(f'{row[0]},{point},False')
```
