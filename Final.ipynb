```python
import pandas as pd
import geopandas as gp
import fiona
from shapely.geometry import shape,mapping, Point, Polygon, MultiPolygon
import csv
```

```python
df = gp.read_file('sites.geojson')
df
```

```python
with open('classicFastPassResults.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    next(readCSV)
    for row in readCSV:
        lat = float(row[14])
        lon = float(row[15]) 
        point = Point(lon, lat)
#         print(point) 
        ans = df['geometry'].contains(point)
#         name = df['disp_name']
        if(sum(ans) != 0):
            
            index = ans[ans == True].index[0]
            TrimblePOI = df.iloc[index].disp_name
            print(f'{row[0]}, {TrimblePOI}, True')
        else:
            print(f'{row[0]}, NOT FOUND, False')
```
